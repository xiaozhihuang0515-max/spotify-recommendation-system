```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("dataset.csv")
df.head()
```

```python
df = df.drop(columns=["Unnamed: 0"])
```

```python
df.describe()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt

df['popularity'].hist(bins=30)
plt.title("Popularity Distribution")
plt.show()
```

```python
df.sort_values('popularity', ascending=False).head(10)
```

```python
df['artists'].value_counts().head(10)
```

```python
df.plot.scatter(x='energy', y='danceability')
plt.title("Energy vs Danceability")
plt.show()
```

```python
df.plot.scatter(x='loudness', y='energy')
plt.show()
```

```python
df.groupby('explicit')['popularity'].mean()
```

```python
df.groupby('explicit')['popularity'].mean().plot(kind='bar')
plt.title("Explicit vs Popularity")
plt.show()
```

```python
top_artists = df['artists'].value_counts().head(10)

top_artists.plot(kind='bar')
plt.title("Top 10 Artists by Number of Songs")
plt.show()
```

```python
df['duration_ms'].hist(bins=40)
plt.title("Song Duration Distribution")
plt.show()
```

```python
df.plot.scatter(x='energy', y='popularity')
plt.title("Energy vs Popularity")
plt.show()
```

```python
df.plot.scatter(x='danceability', y='popularity')
plt.title("Danceability vs Popularity")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```
